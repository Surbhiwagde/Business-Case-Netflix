```python
# Handling the null values
df = df[df['duration'].isnull()==False]
df = df[df['date_added'].isnull()==False]
df = df[df['rating'].isnull()==False]

df['cast'].fillna('Unknown',inplace=True)
df['country'].fillna('Unknown',inplace=True)
df['director'].fillna('Unknown',inplace=True)

df['date_add']=pd.to_datetime(df['date_added'], errors='coerce')
df['month']=df['date_add'].dt.month.fillna(-1)
df['month']=df['month'].astype('int64')
```

## **Unnesting director, cast, country, listed_in columns**

```python
def unnesting(row):
  replicator=0
  show_id=[row[0]]
  type_ = [row[1]]
  title = [row[2]]
  date_added = [row[6]]
  release_year = [row[7]]
  rating = [row[8]]
  duration = [row[9]]
  description = [row[11]]
  director = []
  cast=[]
  country=[]
  listed_in = []
  for i in row[3].split(', '):
    for j in row[4].split(', '):
      for k in row[5].split(', '):
        for l in row[10].split(', '):
          director.append(i)
          cast.append(j)
          country.append(k)
          listed_in.append(l)
          replicator += 1
 
  show_id = show_id * replicator
  type_ = type_ * replicator
  title = title * replicator
  date_added = date_added * replicator
  release_year = release_year * replicator
  rating = rating * replicator
  duration = duration * replicator
  description = description * replicator
  return pd.DataFrame({'show_id':show_id, 'type':type_, 'title':title, 'director':director,
  'cast':cast, 'country':country, 'date_added':date_added,
  'release_year':release_year, 'rating':rating, 'duration':duration,
  'listed_in':listed_in, 'description':description})
```

```python
unnested_list = df.apply(lambda x:unnesting(list(x)),axis=1).to_list()
```

```python
unnested_data = pd.concat(unnested_list,ignore_index=True)
unnested_data.head(10)
```
